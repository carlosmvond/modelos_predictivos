**Que son las librerias?**


En el contexto de Python, las «librerías» (también conocidas como «bibliotecas» o «módulos») son colecciones de funciones, clases y métodos predefinidos que extienden la funcionalidad básica del lenguaje. Estas librerías están escritas en Python o en otros lenguajes de programación y se pueden importar en tus programas para reutilizar el código y acceder a funcionalidades adicionales sin tener que escribir todo desde cero.

* **NumPy:** Esta librería proporciona soporte para operaciones matemáticas y numéricas eficientes en matrices y arreglos. Es la base de muchas otras librerías de Data Science.

* **Pandas:** Pandas se utiliza para la manipulación y análisis de datos estructurados en forma de DataFrames. Facilita la limpieza de datos, la combinación de conjuntos de datos y la realización de operaciones de agregación.

* **Matplotlib y Seaborn:** Estas librerías se utilizan para la visualización de datos, permitiendo a los científicos de datos crear gráficos, histogramas, diagramas de dispersión y más.



```python
# Importamos la biblioteca numpy para operaciones numéricas y matplotlib para gráficos
import numpy as np
import matplotlib.pyplot as plt

# Definimos los parámetros de la distribución normal
# La media (μ) de la distribución normal. Este es el valor alrededor del cual se centran los datos.
media = 0      # Ejemplo de media: 0

# La desviación estándar (σ) de la distribución normal. Mide la dispersión de los datos alrededor de la media.
desviacion = 1 # Ejemplo de desviación estándar: 1

# La cantidad de números aleatorios que deseamos generar. Este valor determina cuántos datos generaremos.
cantidad = 1000 # Ejemplo de cantidad: 1000 números

# Generamos números aleatorios a partir de una distribución normal con los parámetros especificados
# np.random.normal() es la función de numpy que nos permite generar estos números.
# - loc: La media de la distribución.
# - scale: La desviación estándar de la distribución.
# - size: El número total de valores aleatorios a generar.

numeros_aleatorios = np.random.normal(loc=media, scale=desviacion, size=cantidad)

# Imprimimos los primeros 10 números aleatorios generados para tener una vista rápida de los datos
print("Primeros 10 números aleatorios generados:")
print(numeros_aleatorios[:10])
```

# **Estructura de datos: Arrays**

En Python, los arrays (o arreglos) se utilizan para almacenar colecciones de datos. Aunque Python no tiene un tipo de dato llamado "array" en el sentido más estricto, hay varias formas de trabajar con colecciones de datos que funcionan de manera similar a los arrays. Las dos formas más comunes de crear y manipular arrays en Python son mediante el uso de listas y la biblioteca NumPy.

```python
# Creamos un array unidimensional (vector) con números enteros
array_unidimensional = np.array([1, 2, 3, 4, 5])
print("Array Unidimensional:")
print(array_unidimensional)
```

```python
# Creamos un array bidimensional (matriz) con números enteros
array_bidimensional = np.array([[1, 2, 3], [4, 5, 6]])
print("\nArray Bidimensional:")
print(array_bidimensional)

type(array_bidimensional)
```

```python
# Acceder a un elemento en un array bidimensional
# La sintaxis es array[fila, columna]

elemento_bidimensional = array_bidimensional[1, 2]  # Segundo fila, tercer columna

print(elemento_bidimensional)
```

```python

# Operación de suma en un array unidimensional
# Sumamos 10 a cada elemento del array
array_suma = array_unidimensional + 10
print("\nArray unidimensional después de sumar 10 a cada elemento:")
print(array_suma)
```

```python
# Operación de multiplicación en un array bidimensional
# Multiplicamos cada elemento del array bidimensional por 2
array_multiplicacion = array_bidimensional * 2
print("\nArray bidimensional después de multiplicar cada elemento por 2:")
print(array_multiplicacion)
```

```python
# Operación de suma entre dos arrays unidimensionales
# Creamos otro array unidimensional
array_adicional = np.array([5, 4, 3, 2, 1])
array_suma_arrays = array_unidimensional + array_adicional
print("\nSuma entre el array unidimensional original y otro array:")
print(array_suma_arrays)
```

```python
# --- Funciones Útiles ---

# Calcular la media de los elementos en el array unidimensional
media = np.mean(array_unidimensional)
print("\nMedia de los elementos en el array unidimensional:")
print(media)

# Calcular la desviación estándar de los elementos en el array unidimensional
desviacion_estandar = np.std(array_unidimensional)
print("\nDesviación estándar de los elementos en el array unidimensional:")
print(desviacion_estandar)
```

```python

# Parámetros de la distribución normal
media = 0      # Media (mu)
desviacion = 1 # Desviación estándar (sigma)
cantidad = 10000 # Cantidad de números aleatorios a generar

# Generar números aleatorios
numeros_aleatorios = np.random.normal(loc=media, scale=desviacion, size=cantidad)

# Creación del histograma para visualizar la distribución de los números aleatorios generados
# plt.figure() crea una nueva figura para el gráfico. 'figsize' especifica el tamaño de la figura en pulgadas.
plt.figure(figsize=(10, 6))


# plt.hist() crea un histograma de los datos
# - numeros_aleatorios: los datos a graficar
# - bins: número de intervalos en los que se dividen los datos para el histograma
# - density: si True, normaliza el histograma para que la suma de las alturas de las barras sea igual a 1
# - alpha: transparencia de las barras del histograma (0 es completamente transparente, 1 es completamente opaco)
# - color: color de las barras del histograma

plt.hist(numeros_aleatorios, bins=30, density=True, alpha=0.6, color='g')


# Añadimos títulos y etiquetas al gráfico para mayor claridad
plt.title('Histograma de Números Aleatorios con Distribución Normal')
plt.xlabel('Valor')  # Etiqueta del eje x
plt.ylabel('Densidad')  # Etiqueta del eje y

# Añadimos una cuadrícula al gráfico para facilitar la lectura de los valores
plt.grid(True)

# Finalmente, mostramos el gráfico generado
plt.show()
```

El siguiente código agrega líneas verticales en el histograma para marcar el promedio de los datos y los percentiles del 2.5% y 97.5%. Estas líneas ayudan a visualizar el intervalo en el que se encuentra el 95% de los valores, proporcionando una representación clara del rango de credibilidad del 95% para la distribución de los datos.

```python
# Calcular el promedio de los números aleatorios generados
promedio = np.mean(numeros_aleatorios)

# Calcular el intervalo de credibilidad del 95%
# Para una distribución normal, los percentiles 2.5% y 97.5% representan el intervalo de credibilidad del 95%
percentil_inferior = np.percentile(numeros_aleatorios, 2.5)
percentil_superior = np.percentile(numeros_aleatorios, 97.5)


# Creación del histograma para visualizar la distribución de los números aleatorios generados
plt.figure(figsize=(10, 6))

# Crear el histograma
plt.hist(numeros_aleatorios, bins=30, density=True, alpha=0.6, color='g', edgecolor='black')

# Añadir líneas verticales
# Línea vertical para el promedio
plt.axvline(promedio, color='r', linestyle='--', linewidth=2, label=f'Promedio ({promedio:.2f})')

# Líneas verticales para el intervalo de credibilidad
plt.axvline(percentil_inferior, color='b', linestyle='--', linewidth=2, label=f'2.5% Percentil')
plt.axvline(percentil_superior, color='b', linestyle='--', linewidth=2, label=f'97.5% percentil')

# Añadimos títulos y etiquetas al gráfico para mayor claridad
plt.title('Histograma de Números Aleatorios con Distribución Normal')
plt.xlabel('Valor')  # Etiqueta del eje x
plt.ylabel('Densidad')  # Etiqueta del eje y

# Añadimos una leyenda para identificar las líneas verticales
plt.legend()

# Añadimos una cuadrícula al gráfico para facilitar la lectura de los valores
plt.grid(True)

# Finalmente, mostramos el gráfico generado
plt.show()
```
