ArviZ es una biblioteca de Python diseñada para el análisis estadístico y la visualización de modelos probabilísticos. Entre sus muchas funcionalidades, ArviZ proporciona herramientas para calcular intervalos de densidad alta (HDI, por sus siglas en inglés), que son útiles para resumir y visualizar distribuciones de probabilidad.

El siguiente código genera números aleatorios que siguen una distribución normal. Al respecto, se le pide responder:



*  Ejecute el código cambiando la cantidad de números aleatorios que se generan.
*  Ejecute el código considerando un HDI de 50% de probabilidad y un HDI de 10% de probabilidad.
*  Modifique el código de tal manera que se generen números aleatorios utilzando otra distribución (Log-Normal, t-Student, Half-Normal, Cauchy, Beta)








```python
#!pip install arviz  #Por si necesitan instalar arviz

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
import arviz as az


# Parámetros de las distribuciones
mu_normal, sigma_normal = 0, 1  # Media y desviación estándar para la normal


# Simular muestras para la HDI
np.random.seed(42)  # Para reproducibilidad
samples = np.random.normal(mu_normal, sigma_normal, 10000)

# Calcular el intervalo de densidad alta (HDI) para la distribución normal
hdi_normal = az.hdi(samples, hdi_prob=0.95)

# Crear la figura y los subplots
fig, axs = plt.subplots(1, 1, figsize=(10, 6))

# Gráfico para la distribución normal

# Añadir el histograma de las muestras simuladas
axs.hist(samples, bins=50, density=True, alpha=0.3, color='blue', edgecolor='black', label='Histograma de muestras')

# Añadir líneas verticales para el HDI
axs.axvline(hdi_normal[0], color='blue', linestyle='--', label=f'HDI izquierda = {hdi_normal[0]:.2f}')
axs.axvline(hdi_normal[1], color='blue', linestyle='--', label=f'HDI derecha = {hdi_normal[1]:.2f}')

# Línea vertical para la media de la distribución
axs.axvline(mu_normal, color='r', linestyle='--', label=f'Media = {mu_normal}')

# Configuración de los ejes y título
axs.set_title('Distribución Normal con Histograma y HDI', fontsize=14)
axs.set_xlabel('x')
axs.set_ylabel('Densidad')
axs.grid(True)
axs.legend()

# Ajustar el diseño y mostrar la figura
plt.tight_layout()
plt.show()
```
